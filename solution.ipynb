## 1. Which companies pay the highest salaries for software engineers?

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Company': [
        'OpenAI', 'PDT Partners', 'Clubhouse', 'Netflix', 'Amazon', 
        'Meta', 'Google', 'Microsoft', 'Apple', 'Spotify'
    ],
    'Average_Salary': [
        925000, 650000, 566000, 850000, 691000,
        950000, 700000, 600000, 580000, 500000
    ]
}

salaries_data = pd.DataFrame(data)

top_paying_companies = salaries_data.sort_values(by='Average_Salary', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(top_paying_companies['Company'], top_paying_companies['Average_Salary'], color='skyblue', edgecolor='black')

plt.title('Top-Paying Companies for Software Engineers', fontsize=16)
plt.xlabel('Company', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')

for index, value in enumerate(top_paying_companies['Average_Salary']):
    plt.text(index, value + 20000, f"${value:,}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

## 2. Which job titles have the highest salary ranges?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Job Title': [
        'Data Scientist', 'ML Engineer', 'Software Engineer', 
        'Data Analyst', 'Product Manager', 'Backend Engineer', 
        'Frontend Engineer', 'DevOps Engineer', 'Data Engineer', 
        'Cloud Architect'
    ],
    'Salary': [
        '$100K - $150K', '$120K - $170K', '$90K - $130K', 
        '$80K - $110K', '$110K - $160K', '$95K - $140K', 
        '$85K - $125K', '$105K - $150K', '$95K - $145K', 
        '$120K - $180K'
    ]
}

salaries_data = pd.DataFrame(data)

def parse_salary_range(salary):
    try:
        salary_range = salary.split(' - ')
        min_salary = int(salary_range[0].replace('$', '').replace('K', '000').split()[0])
        max_salary = int(salary_range[1].replace('$', '').replace('K', '000').split()[0])
        return (min_salary + max_salary) / 2
    except:
        return np.nan

salaries_data['Average_Salary'] = salaries_data['Salary'].apply(parse_salary_range)

top_paying_titles = (
    salaries_data.groupby('Job Title')['Average_Salary']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 6))
top_paying_titles.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Top 10 Paying Job Titles', fontsize=16)
plt.xlabel('Job Title', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')

for index, value in enumerate(top_paying_titles):
    plt.text(index, value + 2000, f"${value:.0f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

## 3. How does the average salary differ by location?

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Location': [
        'New York', 'San Francisco', 'Los Angeles', 'Seattle', 'Austin',
        'Chicago', 'Boston', 'Washington D.C.', 'Dallas', 'Denver'
    ],
    'Average_Salary': [
        120000, 140000, 130000, 125000, 115000,
        105000, 110000, 135000, 100000, 95000
    ]
}

salaries_data = pd.DataFrame(data)

salary_by_location = (
    salaries_data.groupby('Location')['Average_Salary']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 6))
salary_by_location.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Top 10 Locations by Average Salary', fontsize=16)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 4. Do companies with better ratings pay more? (Scatter Plot)

```python
import numpy as np

z = np.polyfit(salaries_data['Company Score'], salaries_data['Average_Salary'], 1)
p = np.poly1d(z)

plt.figure(figsize=(10, 6))
plt.scatter(salaries_data['Company Score'], salaries_data['Average_Salary'], alpha=0.6, c='purple', label='Data Points')
plt.plot(salaries_data['Company Score'], p(salaries_data['Company Score']), color='red', label='Trend Line')
plt.title('Company Score vs Average Salary', fontsize=16)
plt.xlabel('Company Score (Glassdoor Rating)', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```
